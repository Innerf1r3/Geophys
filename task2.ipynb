```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma_external = 2.9
sigma = 0
R = 10000
h = 20000
delta_sigma = sigma - sigma_external
G = 6.674e-8
```

```python
def V_derevative_for_X(delta_sigma, ksi, eta, dzeta):
    m = delta_sigma * 4/3 * R ** 3
    return G * m * ksi/np.power(((dzeta + h) ** 2 + ksi ** 2 + eta ** 2), 1.5)

def V_derevative_for_Z(delta_sigma, ksi, eta, dzeta):
    m = delta_sigma * 4/3 * R ** 3
    return G * m * (dzeta + h)/np.power(((dzeta + h) ** 2 + (ksi) ** 2 + eta ** 2), 1.5)
```

```python
plt.figure(figsize=(6, 6))

x = np.linspace(-100000, 100000, 1000)
V_derevative_for_X_result = V_derevative_for_X(delta_sigma, x, 0, 0)
V_derevative_for_Z_result = V_derevative_for_Z(delta_sigma, x, 0, 0)

plt.plot(x, V_derevative_for_X_result)
plt.plot(x, V_derevative_for_Z_result)

ax = plt.gca()
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
