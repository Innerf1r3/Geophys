```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma_external = 2.9
sigma = 0
R = 10000
h = 20000
delta_sigma = -sigma
G = 6.674e-8
```

```python
def V_derevative_for_X(delta_sigma, ksi, eta, dzeta):
    m = delta_sigma * 4/3 * R ** 3
    return G * m * ksi/np.power(((dzeta + h) ** 2 + ksi ** 2 + eta ** 2), 0.5)

def V_derevative_for_Z(delta_sigma, ksi, eta, dzeta):
    m = delta_sigma * 4/3 * R ** 3
    return G * m * (dzeta + h)/np.power(((dzeta + h) ** 2 + ksi ** 2 + eta ** 2), 0.5)
```

```python
plt.figure(figsize=(6, 6))

x = np.linspace(-1000, 1000, 1000)
V_derevative_for_X_result = V_derevative_for_X(delta_sigma, x, 0, 0)

plt.plot(x, V_derevative_for_X_result)
```
